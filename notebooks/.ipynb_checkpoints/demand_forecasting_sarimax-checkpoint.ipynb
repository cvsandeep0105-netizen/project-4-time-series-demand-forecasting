## Step 2: Dataset Loading & Initial Exploration

In this step, we load the demand dataset and inspect its structure, size,
and data types to understand the business context before any preprocessing
or forecasting.

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)

# Load dataset
df = pd.read_csv("data/demand_data.csv")

# Preview data
df.head()
```

```python
df.info()
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(
    df['Date'],
    errors='coerce',   # converts bad values to NaT instead of crashing
    dayfirst=True
)
```

```python
df = df.dropna(subset=['Date'])
```

```python
df = df.sort_values('Date')
df.set_index('Date', inplace=True)
```

```python
daily_demand = df['Order_Demand'].groupby(df.index).sum()
daily_demand.head()
```

```python
monthly_demand = daily_demand.resample('M').sum()
monthly_demand.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(monthly_demand)
plt.title("Monthly Demand Trend")
plt.xlabel("Date")
plt.ylabel("Total Demand")
plt.grid(True)
plt.show()
```

## Train‚ÄìTest Split (Time Series)
We split the data chronologically to avoid data leakage.

```python
# Train-test split (last 3 months as test)
train = monthly_demand.iloc[:-3]
test = monthly_demand.iloc[-3:]

print("Train shape:", train.shape)
print("Test shape:", test.shape)
```

## Baseline Forecast (Naive Model)
Assumes future demand equals last observed value.
Used as a benchmark.

```python
# Naive forecast
naive_forecast = test.copy()
naive_forecast[:] = train.iloc[-1]

naive_forecast
```

## Baseline Model Evaluation
RMSE is used to measure forecasting accuracy.

```python
from sklearn.metrics import mean_squared_error
import numpy as np

rmse_naive = np.sqrt(mean_squared_error(test, naive_forecast))
print("Naive Model RMSE:", rmse_naive)
```

## ARIMA Model for Demand Forecasting
ARIMA captures trend and temporal dependencies in time-series data.

```python
!pip install statsmodels
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
# Train-test split (80% train, 20% test)
split_point = int(len(monthly_demand) * 0.8)

train = monthly_demand.iloc[:split_point]
test  = monthly_demand.iloc[split_point:]

print("Train size:", len(train))
print("Test size:", len(test))
```

```python
arima_model = ARIMA(train, order=(1,1,1))
arima_fit = arima_model.fit()

arima_fit.summary()
```
